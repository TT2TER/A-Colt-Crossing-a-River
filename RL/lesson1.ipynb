# 强化学习

### agent&env

状态S作为agent的输入:当前的界面作为状态，木板位置，块位置……

agent做动作action:小木板如何移动

env给agent反馈r

**奖赏值r**

### 四个子要素
- 策略policy 会用决策选取动作
- 奖赏函数 reward fucntion 评价动作获得的回报
- 值函数 value fuction 对当前状态评估  值函数越大，智能体进入这个状态越有利。
- 环境模型

#### 策略
从状态到action的映射
确定性策略（贪婪策略）：选概率最大的动作
$a=\pi(S)$

随机策略：按照概率来，就是在试错
$\pi(a|s)=P[At=a|St=s]$

#### 奖赏函数
把环境中感知到的状态映射为奖赏值，表示该状态内在的可取程度
$r(s,a)=E[R_{t+1}|S_t=s,A_t=a]$
？？

#### 值函数
评价状态的好坏，值函数越大，智能体进入这个状态越有利。
$v_{\pi}(s)=E_{\pi}[R_{t+1}+\gamma R_{t+2}+\gamma^2 R_{t+3}+...|S_t=s]$
？？
值函数是对奖赏的预测

值函数可以改善策略，策略可以改善值函数

#### 环境模型
环境模型是对外界环境的内部运行机制的预测，可以预测下一个状态和奖赏值


### 强化学习的分类
- 基于值的强化学习 value based  可以隐式的学习策略，通过优化值函数来提高性能。
- 基于策略的强化学习 policy based   显式地建模策略，然后通过优化策略来提高性能 没有值函数。走完一个ep后，才能更新策略
- 基于演员-评论员的强化学习 actor-critic based  既有值函数，又有策略函数

- 基于环境模型是否存在
    - 无模型强化学习 model-free  无模型强化学习是指智能体不知道环境的运行机制，只能通过与环境交互来学习。显式学习策略/值函数
    - 有模型强化学习 model-based  有模型强化学习是指智能体知道环境的运行机制，可以通过环境模型来预测下一个状态和奖励值。可以没有值函数，也可以没有策略函数
### 强化学习的上限是人的能力
- 没有训练数据
- 延迟奖励
- 智能体获得自己能力的过程，其实是不断地试错探索（trial-and-error exploration）的过程。探索 （exploration）和利用（exploitation）是强化学习里面非常核心的问题。其中，探索指尝试一些新的动作， 这些新的动作有可能会使我们得到更多的奖励，也有可能使我们“一无所有”；利用指采取已知的可以获得最多奖励的动作，重复执行这个动作，因为我们知道这样做可以获得一定的奖励。因此，我们需要在探 索和利用之间进行权衡，总体趋势，探索变少，利用变多

```python
import gym

class SimpleAgent:
    def __init__(self, env):
        pass
    
    def decide(self, observation): # 决策
        position, velocity = observation
        lb = min(-0.09 * (position + 0.25) ** 2 + 0.03,
                0.3 * (position + 0.9) ** 4 - 0.008)
        ub = -0.07 * (position + 0.38) ** 2 + 0.07
        if lb < velocity < ub:
            action = 2
        else:
            action = 0
        return action # 返回动作

    def learn(self, *args): # 学习
        pass

def play(env, agent, render=False, train=False):
        episode_reward = 0. # 记录回合总奖励，初始值为0
        observation = env.reset() # 重置游戏环境，开始新回合
        while True: # 不断循环，直到回合结束
            if render: # 判断是否显示
                env.render() # 显示图形界面
            action = agent.decide(observation)
            next_observation, reward, done, _ = env.step(action) # 执行动作
            episode_reward += reward # 收集回合奖励
            if train: # 判断是否训练智能体
                agent.learn(observation, action, reward, done) # 学习
            if done: # 回合结束，跳出循环
                break
            observation = next_observation
        return episode_reward # 返回回合总奖励
    

env = gym.make('MountainCar-v0')
print('观测空间 = {}'.format(env.observation_space))
print('动作空间 = {}'.format(env.action_space))
print('观测范围 = {} ~ {}'.format(env.observation_space.low,
        env.observation_space.high))
print('动作数 = {}'.format(env.action_space.n))
env.seed(666) # 设置随机种子，让结果可复现
agent = SimpleAgent(env)
episode_reward = play(env, agent, render=True)
print('回合奖励 = {}'.format(episode_reward))
env.close() # 关闭图形界面
```
