```python
import torch
import torch.nn as nn

m = nn.Softmax(dim=0)
input = torch.randn(2, 3,4,2) #https://pytorch.org/docs/stable/generated/torch.randn.html
print(input.shape)
print(input.dim())
print(input)
# input=input.unsqueeze(dim=0) #升维
print("unsqueeze")
print(input.shape)
print(input.dim())
print(input)
output = m(input)
print(output)
```

dim是tenser的[维度数](https://pytorch.org/docs/stable/generated/torch.Tensor.dim.html#torch.Tensor.dim)，dim=0永远表示最外层的一个维度，dim=1表示从外向内第二层的一个维度，以此类推。dim=-1表示最内层的一个维度，dim=-2表示从内向外第二层的一个维度，以此类推。

randn生成的tensor的维度是由参数个数确定的，第一个参数是dim=0的维度大小，第二个参数是dim=1的维度大小，以此类推。如果只有一个参数，那么生成的tensor是一维的，其dim=0的维度大小就是这个参数的值。

unsqueeze(dim)可以在dim维度上增加一个维度，dim的取值范围是[-(dim+1), dim+1]，dim=0表示在最外层增加一个维度,该维度是由原来的低维构成的，dim=-1表示在最内层增加一个维度，该维度是原来最高维每个元素组成的。

Softmax(dim=a),a的意义和unsqueeze(dim)的意义一样，表示在dim维度上进行softmax操作，dim=0表示在最外层进行softmax操作，dim=-1表示在最内层进行softmax操作。操作的方向便是该维增加的方向。如果该维只有一层`(shape[a]==1)`，那么softmax操作就是把该维所有元素变为1。该维的大小表示softmax求和等与1的元素个数。
